# Fake Job Postings

```python
import os
import numpy as np
import pandas as pd
```

# Paths and Variables

```python
dataset_name = 'fake_job_postings'
```

```python
input_dir = './data'
output_dir = f'./../../processed/{dataset_name}/'
os.makedirs(output_dir, exist_ok=True)
outp_fname = os.path.join(output_dir, f'{dataset_name}.csv')
```

```python
inp_fname = 'fake_job_postings.csv'
```

# Read data into a DataFrame

```python
data = pd.read_csv(os.path.join(input_dir, inp_fname))
print(data.shape)
data.head()
```

# Prepare Data

```python
id_col = "job_id"
# text will be concatenation of company_profile, description, and requirements fields in original data. 
text_col = "company_profile_description_requirements"
target_col = "fraudulent"
```

```python
# We will fill nan values with custom strings
data['company_profile'] = data['company_profile'].fillna("company profile not available")
data['description'] = data['description'].fillna("description not available")
data['requirements'] = data['requirements'].fillna("requirements not available")
```

```python
data[target_col].value_counts()
```

```python
# concatenate the 3 fields. 
data[text_col] = data.apply( lambda row:  str(row['company_profile']) + " " + str(row['description'])
                             + " " + str(row['requirements']) , axis = 1)
```

```python
data.head()
```

# Shuffle Data

```python
# shuffle data
data = data.sample(frac=1, random_state=42)
print(data.shape)
data.head()
```

# Insert Id Column

```python
# insert Id column 
if id_col not in data.columns:
    N = data.shape[0]
    data.insert(0, id_col, np.arange(N))
    print(data.head())
```

```python
data = data[[id_col, target_col, text_col]]
data.head()
```

```python
data[id_col].nunique(), data[target_col].nunique(), data[text_col].nunique()
```

# Utility to Save DF as a zipped file

```python
def save_df_to_zipped_csv(df, ftype=None): 
    if ftype is not None: 
        suffix = f'_{ftype}'
    else: 
        suffix = ''
        
    zipped_f_name = f'{dataset_name}{suffix}.zip'
    archive_f_name = f'{dataset_name}{suffix}.csv'   
    compression_opts = dict(method='zip',
                        archive_name=archive_f_name)      
    df.to_csv(os.path.join(output_dir, zipped_f_name), index=False, compression=compression_opts )
```

# Save Main Data File

```python
# # save original file as csv
# data.to_csv(outp_fname, index=False)

# save as zipped file 
save_df_to_zipped_csv(data)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
test_size = 0.1

data_train, data_test = train_test_split(data, test_size=test_size, random_state=42)


data_test_key = data_test[[id_col, target_col]].copy()
data_test = data_test.drop(columns=[target_col])
print(data_train.shape, data_test.shape, data_test_key.shape)

# # Save original files as csv
# data_train.to_csv(outp_train_fname, index=False)
# data_test.to_csv(outp_test_fname, index=False)
# data_test_key.to_csv(outp_test_key_fname, index=False)
```

```python
# zip files
save_df_to_zipped_csv(data_train, "train")
save_df_to_zipped_csv(data_test, "test")
save_df_to_zipped_csv(data_test_key, "test_key")
```
